Problem 1: Mobile Recharge Validation System

Problem Statement--
Create a function to validate a mobile recharge request.


```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]

    if amount < 50:
        print("âŒ Recharge amount must be at least â‚¹50.")
        return False
    elif amount not in valid_plans:
        print("âŒ Invalid plan selected.")
        print("Available plans: 199, 299, 399, 599")
        return False
    else:
        print("âœ… Recharge successful!")
        return True


# Retry system using while loop
while True:
    try:
        user_amount = int(input("Enter recharge amount: â‚¹"))

        if validate_recharge(user_amount):
            break  # Exit loop if valid

        print("Please try again.\n")

    except ValueError:
        print("âŒ Please enter a valid numeric amount.\n")
```

Problem 2: Inventory Reorder Alert System

Problem Statement--
Given a dictionary of products and stock quantities:
Print Reorder Alert if stock < 15
Print Stock OK otherwise


```python
def check_inventory(stock_data):
    for product, quantity in stock_data.items():
        if quantity < 15:
            print(f"âš ï¸ {product}: Reorder Alert (Stock = {quantity})")
        else:
            print(f"âœ… {product}: Stock OK (Stock = {quantity})")


# Sample dictionary of products
inventory = {
    "Rice": 25,
    "Wheat": 10,
    "Sugar": 18,
    "Oil": 12,
    "Salt": 30
}

check_inventory(inventory)
```

Problem 3: Student Result Processing System

Problem Statement--
Create a function that:
Accepts a list of student marks Calculates average marks Prints Pass if average â‰¥ 50 else Fail




```python
def process_result(marks):
    total = 0

    # Loop to calculate total marks
    for mark in marks:
        total += mark

    average = total / len(marks)

    if average >= 50:
        return f"Average = {average:.2f} â†’ Pass"
    else:
        return f"Average = {average:.2f} â†’ Fail"


# Example usage
student_marks = [60, 45, 55, 70, 40]

result = process_result(student_marks)
print(result)
```

Problem 4: Cab Fare Estimator with Retry Option

Problem Statement--Create a cab fare calculator:
Base fare = â‚¹50,â‚¹12 per km,Peak hour â†’ 25% extra,Allow retry using while loop



```python
def calculate_fare(distance, peak):
    base_fare = 50
    per_km_rate = 12

    fare = base_fare + (distance * per_km_rate)

    if peak.lower() == "yes":
        fare += fare * 0.25   # 25% extra during peak hours

    return fare


# Retry system using while loop
while True:
    try:
        distance = float(input("Enter distance in km: "))
        peak_hour = input("Is it peak hour? (yes/no): ")

        total_fare = calculate_fare(distance, peak_hour)

        print(f"ðŸš– Total Fare: â‚¹{total_fare:.2f}")
        break   # Exit loop after successful calculation

    except ValueError:
        print("âŒ Invalid input. Please enter numeric distance.\n")
```

Problem 5: Employee Attendance Eligibility Checker

Problem Statement--
Create a function that:
Accepts attendance list ("P", "A"), Calculates attendance percentage,Returns Eligible if â‰¥ 75% else Not Eligible



```python
def check_attendance(attendance_list):
    present_count = 0

    # Loop to count "P"
    for status in attendance_list:
        if status.upper() == "P":
            present_count += 1

    total_days = len(attendance_list)
    percentage = (present_count / total_days) * 100

    if percentage >= 75:
        return f"Attendance = {percentage:.2f}% â†’ Eligible"
    else:
        return f"Attendance = {percentage:.2f}% â†’ Not Eligible"


# Example usage
attendance = ["P", "A", "P", "P", "A", "P", "P", "P"]

result = check_attendance(attendance)
print(result)
```

Problem 6: Password Strength Checker

Problem Statement--
Create a function to check password strength.
Rules:
Minimum length = 8, Must contain at least one digit, Must contain at least one special character (@#$)


```python
def check_password_strength(password):
    has_digit = False
    has_special = False

    # Minimum length check
    if len(password) < 8:
        return "âŒ Weak Password (Minimum length is 8)"

    # Loop through each character
    for char in password:
        if char.isdigit():
            has_digit = True
        if char in "@#$":
            has_special = True

    # Conditions check
    if has_digit and has_special:
        return "âœ… Strong Password"
    else:
        return "âŒ Weak Password (Must contain at least one digit and one special character @#$)"


# Example usage
user_password = input("Enter password: ")
result = check_password_strength(user_password)
print(result)
```
